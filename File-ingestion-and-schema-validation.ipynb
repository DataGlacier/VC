```python
import csv
import os
import gzip

# Function to get file size in MB
def get_file_size(file_path):
    return os.path.getsize(file_path) / (1024 * 1024)  # Convert to MB

# Function to calculate the total number of rows and columns in a CSV file
def get_csv_dimensions(file_path):
    with open(file_path, "r") as file:
        csv_reader = csv.reader(file)
        num_rows = sum(1 for row in csv_reader)
        file.seek(0)  # Reset file pointer
        num_columns = len(next(csv_reader))
    return num_rows, num_columns

# Define the folder and file names
folder_name = "DataSets-main"
file_name1 = "City.csv"
file_name2 = "Customer_ID.csv"
file_name3 = "Transaction_ID.csv"
file_name4 = "Cab_Data.csv"

# Construct the full file paths
downloads_path = os.path.expanduser("~" + "/Downloads")  # Get the path to the Downloads folder
file_path1 = os.path.join(downloads_path, folder_name, file_name1)
file_path2 = os.path.join(downloads_path, folder_name, file_name2)
file_path3 = os.path.join(downloads_path, folder_name, file_name3)
file_path4 = os.path.join(downloads_path, folder_name, file_name4)

# Initializing empty lists to store data from each file
data1, data2, data3, data4 = [], [], [], []

# Read data from City.csv
with open(file_path1, mode="r") as file:
    csv_reader = csv.DictReader(file)
    data1 = [row for row in csv_reader]

# Read data from Customer_ID.csv
with open(file_path2, mode="r") as file:
    csv_reader = csv.DictReader(file)
    data2 = [row for row in csv_reader]

# Read data from Transaction.csv
with open(file_path3, mode="r") as file:
    csv_reader = csv.DictReader(file)
    data3 = [row for row in csv_reader]

# Read data from Cab_Data.csv
with open(file_path4, mode="r") as file:
    csv_reader = csv.DictReader(file)
    data4 = [row for row in csv_reader]
```

```python
# Print the number of columns and rows in each file
print("Number of columns and rows in each file:")
print(f"City.csv: Columns - {len(data1[0])}, Rows - {len(data1)}")
print(f"Customer_ID.csv: Columns - {len(data2[0])}, Rows - {len(data2)}")
print(f"Transaction.csv: Columns - {len(data3[0])}, Rows - {len(data3)}")
print(f"Cab_Data.csv: Columns - {len(data4[0])}, Rows - {len(data4)}")

# Combine appropriate files to find the use of pink cabs instead of yellow cabs
pink_cabs = 0
yellow_cabs = 0

for row in data4:
    if row["Company"] == "Pink Cab":
        pink_cabs += 1
    elif row["Company"] == "Yellow Cab":
        yellow_cabs += 1

print(f"Use of Pink Cabs: {pink_cabs} trips")
print(f"Use of Yellow Cabs: {yellow_cabs} trips")

# Find the use of cabs in each named city
city_cab_counts = {}

for row in data4:
    city = row["City"]
    if city in city_cab_counts:
        city_cab_counts[city] += 1
    else:
        city_cab_counts[city] = 1

print("Use of Cabs in Each Named City:")
for city, count in city_cab_counts.items():
    print(f"{city}: {count} trips")

# Write the results to a pipe-separated text file in gz format
output_file = "cab_usage.txt.gz"
with gzip.open(output_file, "wt") as file:
    writer = csv.writer(file, delimiter="|")
    writer.writerow(["City", "Trips"])
    for city, count in city_cab_counts.items():
        writer.writerow([city, count])

print(f"Results written to {output_file}")

# Get the file size, total number of rows, and total number of columns for each file
file_summary = {}

# For City.csv
num_rows_city, num_columns_city = get_csv_dimensions(file_path1)
file_size_city = get_file_size(file_path1)

file_summary["City.csv"] = {
    "Total Rows": num_rows_city,
    "Total Columns": num_columns_city,
    "File Size (MB)": file_size_city
}

# For Customer_ID.csv
num_rows_customer_id, num_columns_customer_id = get_csv_dimensions(file_path2)
file_size_customer_id = get_file_size(file_path2)

file_summary["Customer_ID.csv"] = {
    "Total Rows": num_rows_customer_id,
    "Total Columns": num_columns_customer_id,
    "File Size (MB)": file_size_customer_id
}

# For Transaction.csv
num_rows_transaction, num_columns_transaction = get_csv_dimensions(file_path3)
file_size_transaction = get_file_size(file_path3)

file_summary["Transaction.csv"] = {
    "Total Rows": num_rows_transaction,
    "Total Columns": num_columns_transaction,
    "File Size (MB)": file_size_transaction
}

# For Cab_Data.csv
num_rows_cab_data, num_columns_cab_data = get_csv_dimensions(file_path4)
file_size_cab_data = get_file_size(file_path4)

file_summary["Cab_Data.csv"] = {
    "Total Rows": num_rows_cab_data,
    "Total Columns": num_columns_cab_data,
    "File Size (MB)": file_size_cab_data
}

# Print the file summary
print("File Summary:")
for file_name, details in file_summary.items():
    print(f"File: {file_name}")
    print(f"  Total Rows: {details['Total Rows']}")
    print(f"  Total Columns: {details['Total Columns']}")
    print(f"  File Size (MB): {details['File Size (MB)']:.2f} MB")
    print()
```
